    * $x_1$ is distributed as $N(0,1)$

* Find the eigenvectors and eigenvalues of the covariance matrix of the dataset
* Find the eigenvectors and eigenvalues using SVD. Check that the two procedures yield to same result

```python

N = 1000
x1 = rnd.randn(N)
x2 = x1 + 3 * rnd.randn(N)
x3 = 2 * x1 + x2
features = ['x1', 'x2', 'x3']
dataset = pd.DataFrame(numpy.vstack((x1,x2,x3)).T, columns=features)




cov = np.cov(dataset.values.T)
spect_cov, V_cov = linalg.eig(cov)
spect_cov = np.real_if_close(spect_cov)
V_cov = V_cov.T
sorting = np.argsort(spect_cov)   
spect_cov = spect_cov[sorting][::-1] 
V_cov = V_cov[sorting][::-1]         


#svd
U_SVD, spect_SVD, V_SVD = linalg.svd(dataset.values.T)
spect_SVD = numpy.real_if_close(spect_SVD**2 / (N-1))
U_SVD = U_SVD.T


check = True
for i in range(V_cov.shape[0]):
    check = check and ( numpy.allclose(V_cov[i], U_SVD[i], 1e-2, 1e-10) or numpy.allclose(V_cov[i], -U_SVD[i], 1e-2, 1e-10) )
print(f"Check : {check}\n")


percentage = spect_cov / cov.trace()
print(f"variability: {percentage}")
```
