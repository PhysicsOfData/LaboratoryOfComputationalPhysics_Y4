1\. **2D minimization of a six-hump camelback function**

$$f(x,y) = \left(4-2.1x^2+\frac{x^4}{3} \right) x^2 +xy + (4y^2 -4)y^2$$

has multiple global and local minima.

- Find the global minima of this function
- How many global minima are there, and what is the function value at those points?
- What happens for an initial guess of $(x, y) = (0, 0)$?

Hints:

* Variables can be restricted to $-2 < x < 2$ and $-1 < y < 1$.
* Use `numpy.meshgrid()` and `pylab.imshow()` to find visually the regions.
* Use `scipy.optimize.minimize()`, optionally trying its optional arguments.

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize,stats
import pandas as pd

%matplotlib inline
````

````python
def f(parameters):
    x,y = parameters 
    return (4-2.1*x**2+x**4/3)*x**2+x*y+(4*y**2-4)*y**2
x = np.linspace(-2, 2, 100)
y = np.linspace(-1, 1, 100)

X, Y = np.meshgrid(x, y, sparse=True)
plt.figure()
plt.imshow(f([X,Y]),extent = [-2,2,-1,1], origin = 'lower')
plt.colorbar()
plt.show()

initial_guess = ([0, -0.75],[0,0.75])
result1 = optimize.minimize(f, initial_guess[0])
result2 = optimize.minimize(f, initial_guess[1])
if result1.success:
    fitted_params = result1.x
    print("We have a global minimum in ",fitted_params, "and the value of f at this point is:", f(fitted_params))
if result2.success:
    fitted_params = result2.x
    print("We have a global minimum in ",fitted_params, "and the value of f at this point is:", f(fitted_params))

else:
    raise ValueError(result.message)
wrong_guess = ([0,0])
result = optimize.minimize(f,wrong_guess)
fitted_params3 = result.x
print("With the initial guess [0,0] we find a minimum for",fitted_params3, "and the value of f at this point is:", f(fitted_params3))

````

2\. **Curve fitting of temperature in Alaska** 

The temperature extremes in Alaska for each month, starting in January, are given by (in degrees Celcius):

max:  `17,  19,  21,  28,  33,  38, 37,  37,  31,  23,  19,  18`

min: `-62, -59, -56, -46, -32, -18, -9, -13, -25, -46, -52, -58`

* Plot these temperatures.
* Find a suitable a function that can describe min and max temperatures. 
* Fit this function to the data with `scipy.optimize.curve_fit()`.
* Plot the result. Is the fit reasonable? If not, why?
* Is the time offset for min and max temperatures the same within the fit accuracy?

````python
max_ = np.array([17,19,21,28,33,38,37,37,31,23,19,18])
min_ = np.array([-62,-59,-56,-46,-32,-18,-9,-13,-25,-46,-52,-58])
month = np.linspace(1,12,12)

def sin(x, A, nu, y_offset):
    return y_offset+ A * np.sin(nu * x) # sine function with amplitude A and angular frequency
def gauss(x, mean, sigma, offset, amplitude):
    return amplitude * stats.norm.pdf(x, mean, sigma) + offset

fig, (ax1, ax2) = plt.subplots(nrows = 1, ncols = 2, figsize = (20, 10))
popt_max_sin, pcov_max_sin = optimize.curve_fit(sin, month, max_, full_output=False)
popt_min_sin, pcov_min_sin = optimize.curve_fit(sin, month, min_,[2,0,0])
print(pcov_min_sin)

ax1.plot(month,min_,color = 'blue',marker = 'o',label = 'Minimum')
ax1.plot(month,max_,color = 'red',marker = 'o',label = 'Maximum')
ax1.plot(month, sin(month, *popt_max_sin), label = 'max T')
ax1.plot(month, sin(month, *popt_min_sin), label = 'min T')
ax1.set_title('Fit with Sine')
ax1.set_xlabel('Month')
ax1.set_ylabel('Temperature')
ax1.legend(loc='best')

popt_max_gauss, pcov_max_gauss = optimize.curve_fit(gauss, month, max_, full_output=False)
popt_min_gauss, pcov_min_gauss = optimize.curve_fit(gauss, month, min_,[7.,2.,-60.0, -10])


ax2.plot(month,min_,color = 'blue',marker = 'o',label = 'Minimum')
ax2.plot(month,max_,color = 'red',marker = 'o',label = 'Maximum')
ax2.plot(month, gauss(month, *popt_max_gauss), label = 'max T')
ax2.plot(month, gauss(month, *popt_min_gauss), label = 'min T')
ax2.set_title('Fit with Gaussian')
ax2.set_xlabel('Month')
ax2.set_ylabel('Temperature')
ax2.legend(loc='best')

print("Both fits are rasonable")
````

The feature name "residual" contains the residuals (defined as $y_i - \hat{y}_i$) of a linear regression as a function of the independent variable "distances".

- Considering only the "residual" feature, create an histogram with the appropriate binning and display it.
- Set the appropriate Poisson uncertainty for each bin (thus, for each bin, $\sigma_i = \sqrt{n_i}$, where $n_i$ is the number of entries in each bin)
- By looking at the distribution of the residuals, define an appropriate function and fit it to the histogram of the residuals
- Perform a goodness-of-fit test. Is the p-value of the fit satisfactory?

````python
#!wget https://www.dropbox.com/s/3uqleyc3wyz52tr/residuals_261.pkl -P data/
````

````python
File = np.load("data/residuals_261.pkl", allow_pickle = True)
from matplotlib.ticker import MaxNLocator
data = File.tolist()
Data = pd.DataFrame(data)
df = Data['residuals']
df = df[abs(df) < 1.5]
n_bins = 50
counts,binEdges = np.histogram(df,bins=n_bins)
Mid_bins = 1/2 * (binEdges[1:] + binEdges[:-1])

def gauss(x, mean, sigma, offset, amplitude):
    return amplitude * stats.norm.pdf(x, mean, sigma) + offset

bincenters = 0.5*(binEdges[1:]+binEdges[:-1])
meanStd = np.sqrt(counts)
popt_gauss, pcov_gauss = optimize.curve_fit(gauss, bincenters, counts, full_output=False,p0 = [0.1,0.5,0,1500])
x = np.linspace(int(min(df)),int(max(df)),1000)

fig, ax1 = plt.subplots(figsize=(10, 6))
ax1.hist(df, n_bins, histtype='bar', color='purple', alpha = 0.6)

ax1.yaxis.set_major_locator(MaxNLocator(integer = True))
ax1.errorbar(Mid_bins,counts,yerr = (np.sqrt(counts)), ecolor = 'r',fmt = '.k' )
ax1.plot(x, gauss(x, *popt_gauss), label = 'fitted curve')

chi_sq = np.sum(((counts - gauss(Mid_bins,popt_gauss[0],popt_gauss[1],popt_gauss[2],popt_gauss[3]))/meanStd)**2)
print('The chi square is:', chi_sq)
dof = n_bins - 1
p_value = 1. - stats.chi2.cdf(chi_sq,dof)
print('The p-value is:',p_value)
print('The p-value of the fit is not satisfactory because it is equal to zero')
#ax1.xlim(-5,5)
````

4\. **Temperatures in Munich**

Get the following data file:

```bash
https://www.dropbox.com/s/7gy9yjl00ymxb8h/munich_temperatures_average_with_bad_data.txt
```

which gives the temperature in Munich every day for several years.


Fit the following function to the data:

$$f(t) = a \cos(2\pi t + b)+c$$

where $t$ is the time in years.

- Make a plot of the data and the best-fit model in the range 2008 to 2012.

   - What are the best-fit values of the parameters?

   - What is the overall average temperature in Munich, and what are the typical daily average values predicted by the model for the coldest and hottest time of year?

   - What is the meaning of the $b$ parameter, and what physical sense does it have?


- Now fit the data with the function $g(x)$, which has 1 more parameter than $f(x)$.
$$g(x) = a \cos(2\pi b t + c)+d$$
   - What are the RSS for $f(x)$ and $g(x)$?
   - Use the Fisher F-test to determine whether the additional parameter is necessary.

````python
 #!wget https://www.dropbox.com/s/7gy9yjl00ymxb8h/munich_temperatures_average_with_bad_data.txt
````

````python
data = pd.DataFrame(pd.read_csv("munich_temperatures_average_with_bad_data.txt", header = None, delimiter = " ", names = ["years","temperatures"]))


def f(t,a,b,c):
    return a*np.cos(2*np.pi*t+b)+c

def g(t,a,b,c,d):
    return a*np.cos(2*np.pi*t*b+c)+d

data = data[data['years'] >= 2008]
data = data[data['years'] <= 2012]
x = np.array(data['years'])
y = np.array(data['temperatures'])
x_plot = np.linspace(2008,2012,100)
plt.plot(x,y,marker = '+', ls = " ")
plt.ylim((-20,30))
popt,pcov = optimize.curve_fit(f,x,y,p0=[10,3,10],full_output = False)
popt_g,pcov_g = optimize.curve_fit(g,x,y,full_output = False)
plt.plot(x_plot, f(x_plot, popt[0],popt[1],popt[2]), label = 'fitting curve f(x)', color = 'red')
plt.plot(x_plot,g(x_plot, popt_g[0],popt_g[1],popt_g[2],popt_g[3]), label = 'fitting curve g(x)', color = 'green')
plt.title('Temperatures in Munich 2008-2012')
plt.ylabel('Temperatures (C)')
plt.xlabel('Years')
plt.legend()
plt.show()


def Ftest(ssr_1, ssr_2, ndof_1, ndof_2, nbins, verbose=False):
    F = ((ssr_1 - ssr_2)/(ndof_2 - ndof_1)) / (ssr_2/(nbins - ndof_2))
    CL = 1. - stats.f.cdf(F, ndof_2 - ndof_1, nbins - ndof_2)
    if verbose: print("CL: %.3f" % CL, ", additional parameter necessary:", "YES" if CL < 0.10 else "NO")
    return CL

RSSf,RSSg = 0,0
for i in range(len(x)):
    RSSf += pow(y[i]- f(x[i], popt[0],popt[1],popt[2]),2)
    RSSg += (y[i]- g(x[i], popt_g[0],popt_g[1],popt_g[2],popt_g[3]))**2
    

print("The best fit values are:", '\n a = ',popt[0] ,'+-',pcov[0,0],',\n b = ', popt[1] ,'+-', pcov[1,1],',\n c = ', popt[2] ,'+-', pcov[2,2])
print('Average temperature: %.2fC'%(popt[2]))
print('Hottest average temperature: %.2fC'%(popt[0]+popt[2]))
print('Coldest average temperature: %.2fC'%(popt[0]-popt[2]))
print('b is the phase of the sinusoid function that aligns the maxima with the hottest temperatures')
print("\nRSS of f(x):",RSSf)
print("RSS of g(x):",RSSg)
doff = 3-1
dofg = 4-1
comp = Ftest(RSSf,RSSg, doff, dofg, len(x), verbose=True)
````
