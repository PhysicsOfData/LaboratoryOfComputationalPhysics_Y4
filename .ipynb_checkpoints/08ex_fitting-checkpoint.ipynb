1\. **2D minimization of a six-hump camelback function**

$$f(x,y) = \left(4-2.1x^2+\frac{x^4}{3} \right) x^2 +xy + (4y^2 -4)y^2$$

has multiple global and local minima.

- Find the global minima of this function
- How many global minima are there, and what is the function value at those points?
- What happens for an initial guess of $(x, y) = (0, 0)$?

Hints:

* Variables can be restricted to $-2 < x < 2$ and $-1 < y < 1$.
* Use `numpy.meshgrid()` and `pylab.imshow()` to find visually the regions.
* Use `scipy.optimize.minimize()`, optionally trying its optional arguments.

````python
from scipy.optimize import minimize 
f = lambda x:((4 - 2.1*x[0]**2 + x[0]**4 / 3) * x[0]**2 + x[0] * x[1]
            + (4*x[1]**2 - 4) * x[1] **2)
bnds = ((-2, 2),(-2, 2))
x = np.linspace(-2.0, 2.0)
y = np.linspace(-1.0, 1.0)
x0, x1 = np.meshgrid(x, y)
z = f([x0, x1])
````

````python
from matplotlib import cm
x_global_min = minimize(f, (2,1), bounds=bnds) #using an initail guess at the lower bounds 
x_global_min1 = minimize(f, (-2,-1), bounds= bnds) #using an initail guess at the upper bounds 

plt.figure(figsize=(15,7))
col = plt.imshow(z, extent=[-2, 2, -1, 1],cmap=cm.coolwarm)
plt.scatter(x_global_min.x[0],x_global_min.x[1], color='pink')
plt.scatter(x_global_min1.x[0],x_global_min1.x[1], color='white')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('2D Six-hump Camelback function with 2 global minnma');
plt.colorbar(col);
````

````python

from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(15,7))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(x0, x1, z, rstride=1, cstride=1, cmap=cm.coolwarm,
                        linewidth=0, antialiased=False)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')
ax.set_title('3D Six-hump Camelback function');

print(f"The two global minima are: {x_global_min.x} and {x_global_min1.x}")
print(f"The function value is: {x_global_min.fun}")
````

````python

x_min = minimize(f, (0,0), bounds=bnds)
print(f"The minima is: {x_min.x}")
print(f"The function value is: {x_min.fun}")
plt.figure(figsize=(15,7))
plt.imshow(z, cmap=cm.coolwarm, extent=[-2, 2, -1, 1])
plt.colorbar(col)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('2D Six-hump Camelback function a minnma at the orgin');
plt.scatter(x_min.x[0],x_min.x[1], color='white');
````

2\. **Curve fitting of temperature in Alaska** 

The temperature extremes in Alaska for each month, starting in January, are given by (in degrees Celcius):

max:  `17,  19,  21,  28,  33,  38, 37,  37,  31,  23,  19,  18`

min: `-62, -59, -56, -46, -32, -18, -9, -13, -25, -46, -52, -58`

* Plot these temperatures.
* Find a suitable a function that can describe min and max temperatures. 
* Fit this function to the data with `scipy.optimize.curve_fit()`.
* Plot the result. Is the fit reasonable? If not, why?
* Is the time offset for min and max temperatures the same within the fit accuracy?

````python
max_temp= [17,  19,  21,  28,  33,  38, 37,  37,  31,  23,  19,  18]
min_temp= [-62, -59, -56, -46, -32, -18, -9, -13, -25, -46, -52, -58]
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
plt.plot(months, max_temp, 'ro', label = 'max')
plt.plot(months, min_temp, 'b^', label = 'min')
plt.xlabel('Month')
plt.ylabel('minimum and maxmum temperature')
plt.legend(loc='best')
plt.title('The minimum and maxmum temperatures in Alaska city');
````

````python
from scipy.optimize import curve_fit
def fun(t, A, phi, avg):
    return A * np.cos((t + phi) * 2 * np.pi / t.max()) + avg
#avg = avarage of min and max of respective temps
months = np.arange(12)
#p0=initial guess for the amplitude A, phi and the avarage repestively. 
max_popt, max_pcov = curve_fit(fun, months, max_temp, p0=(10, 0, 28))
min_popt, min_pcov = curve_fit(fun, months, min_temp, p0=(20, 0, -36))
print(f"The optimal parameters for the maximum tempeature:\n{max_popt}")
print(f"Covariance matrix for the max temperature:\n{max_pcov}\n")
print(f"The optimal parameters for the minimum tempeature:\n{min_popt}")
print(f"covariance matrix for the min temperature::\n{min_pcov}")
````

````python
days = np.linspace(0, 12, num=365)
plt.figure(figsize=(5,3), dpi=150)
plt.plot(months, max_temp, 'ro')
plt.plot(days, fun(days, * max_popt), 'r-',label = 'max_fit')
plt.plot(months, min_temp, 'b^')
plt.plot(days, fun(days, * min_popt), 'b--', label = 'min_fit')
plt.xlabel('Month')
plt.legend(loc=(.80, 0.85), fontsize=7)
plt.title('The min and max temperatures in Alaska city')
plt.ylabel('Temperature ($^\circ$C)');
````

- Now fit the data with the function $g(x)$, which has 1 more parameter than $f(x)$.
$$g(x) = a \cos(2\pi b t + c)+d$$

   - Use the Fisher F-test to determine whether the additional parameter is necessary.

````python
url = 'munich_temperatures_average_with_bad_data.txt'
data = pd.read_csv(url, delimiter = ' ', header=None)
new_data=data.loc[(data[0]>=2008) & (data[1]<=2012)]

def fun_mun(t, a, b, c):
    return a * np.cos(2*np.pi*t + b) + c

popt_mun,pcov_mun = curve_fit(fun_mun,new_data[0] ,new_data[1], p0=(100, 0, 10))
print(f"The optimal parameters for the tempeature:\n{popt_mun}\n")
print(f"Covariance matrix for the temperature:\n{pcov_mun}")

````
