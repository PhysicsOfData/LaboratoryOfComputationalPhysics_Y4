1\. **Hurricanes per Year**

The number of hurricanes in 2005 was 15. The historic average is 6.3. Is this number signficantly different?
- Assume the number of hurricanes is random, i.e. follows the Poisson distribution.
- Assume as statistically significant a probability that has a Z score of 3 or larger with respect a normal distribution.

**Hint**: compute the probability that in a single year are observed 15 or more hurricances.

````python
#the probablity of 15 hurricanes in the 2005
mu =6.3
k=15
sigma = np.sqrt(mu)
pmf = stats.poisson.pmf(k, mu)
pmf = np.round(pmf, 5)
#the probablity of more hurricanes in the year
pmf_new = (1 - pmf)*100

print(f"The Probability that in a single year hurricanes observed are more hurricances: {round(pmf_new, 2)}%" )

#Z-value 
Z = abs(mu - k) / sigma
Z_score = round(Z, 4)
print("The Z_score of hurricanes in 2005:", Z_score)

if Z_score >= 3:
    print(f"The number of hurricanes in 2005 make a statistically significant different since {Z_score} is greater than 3")
else:
    print(f"The number of hurricanes in 2005 do not make a statistically significant different since {Z_score} is less than 3")

    
````

2\. **Pairwise t-test**

In an experiment, a group of 10 individuals agreed to participate in a study of blood pressure changes following exposure to halogen lighting. Resting systolic blood pressure was recorded for each individual. The participants were then exposed to 20 minutes in a room lit only by halogen lamps. A post-exposure systolic blood pressure reading was recorded for each individual. The results are presented in the following data set:

```python
pre = np.array([120, 132, 120, 110, 115, 128, 120, 112, 110, 100])
post = np.array([140, 156, 145, 130, 117, 148, 137, 119, 127, 135])
```

Determine whether the change in blood pressures within our sample was statistically significant.

**Hint:**
in this case, the Student's $t$-test should be performed to compare the two datasets.
Use the following test statistics:

$$T = \frac{\bar{x}_1 - \bar{x}_2}{\sigma \sqrt{\frac{2}{n}}}$$

and 

$$\sigma = \sqrt{\frac{\sigma_1^2 + \sigma_2^2}{2}}$$

````python
pre = np.array([120, 132, 120, 110, 115, 128, 120, 112, 110, 100])
post = np.array([140, 156, 145, 130, 117, 148, 137, 119, 127, 135])

pre_mu = np.mean(pre)
post_mu = np.mean(post)
n=10
pre_sigma  = np.sqrt(np.var(pre, ddof=1)) 
post_sigma = np.sqrt(np.var(post, ddof=1))
sigma = np.sqrt((pre_sigma**2+post_sigma**2)/2)

T = (pre_mu -post_mu)/(sigma*np.sqrt(2/n))
print("The T_score is: ",round(T, 4))

pvalue = stats.t.cdf(T, n - 1) + (1. - stats.t.cdf(-T, n - 1))
print("p-value is:", round(pvalue, 4))

# To check the confidence interval / degree of freedom using the upper and lower quantiles
alpha1 = stats.t.ppf(.025, n - 1)
alpha2 = stats.t.ppf(.975, n - 1)
print("The degree of freedom for the lower quantile is: ", round(alpha1, 4))
print("The degree of freedom for the lower quantile is: ", round(alpha2, 4))

if alpha1 <= T <= alpha2:
    print("The T value is within the confidence interval therefore, the change in blood pressure was statistically significant on the sample")
else:
    print("The T value is outside the confidence interval therefore, the change in blood pressure had no statistically significant on the sample ")
````

3\. **FFT of a simple dataset**

Perform a periodicity analysis on the lynxs-hares population, i.e. determine what is the period of the population of these animals.

````python
file = np.loadtxt('populations.txt')
df= pd.DataFrame(file, columns= ["year", "hares", "lynxes", "carrots "])
df.head()
````

````python
fig= plt.figure(figsize=(10,5))
plt.plot(df['year'], df['hares'], label='Hares')
plt.plot(df['year'], df['lynxes'], label='lynxes')
plt.xlabel("Years")
plt.ylabel("Population")
plt.legend();
````

````python

# periodicity analysis on the Hares
hs_fft = fftpack.fft(np.array(df['hares']))
hs_power = np.abs(hs_fft)
hs_freq = fftpack.fftfreq(df['hares'].size, d=1.)

#Find the peak frequency: focus only on only the positive frequencies
h_freq = hs_freq[np.where(hs_freq > 0)]
h_power= hs_power[np.where(hs_freq > 0)]

# find the max of freqs array for the hares
h_peak_freq = h_freq[h_power.argmax()]
h_peak_power = h_power[h_power.argmax()]

print("Hares peak frequency:", h_peak_freq)
print("Hares peak period:", 1/h_peak_freq)


# periodicity analysis on the lynxes
ls_fft = fftpack.fft(np.array(df['lynxes']))
ls_power = np.abs(ls_fft)
ls_freq = fftpack.fftfreq(df['lynxes'].size, d=1.)

# Find the peak frequency: focus only on only the positive frequencies
l_freq = ls_freq[np.where(ls_freq > 0)]
l_power = ls_power[np.where(ls_freq > 0)]

# find the max of freqs array for the lynxes
l_peak_freq = l_freq[l_power.argmax()]
l_peak_power = l_power[l_power.argmax()]

print("Lynxes peak frequency is:", l_peak_freq)
print("Lynxes peak period is: ", 1/l_peak_freq)
````

````python
fig= plt.figure(figsize=(10,5))
plt.plot(hs_freq, hs_power,  label='hares FFT')
plt.plot(hs_freq, ls_power,  label='lynxes FFT')
plt.xlabel("Frequency")
plt.ylabel("Power");
````

4\. **FFT of an image**

Write a filter that removes the periodic noise from the `moonlanding.png` image by using a 2-dimensional FFT.

* Import the image as a 2D numpy array using `plt.imread("moonlanding.png")`. Examine the image with `plt.imshow()`, which is heavily contaminated with periodic noise.
* Check the documentation of the `scipy.fftpack` package, and find the method that performs a 2D FFT. Plot the spectrum (Fourier transform of) the image. **Hint**: use `LogNorm` to plot the colors in log scale:
```Python
from matplotlib.colors import LogNorm
plt.imshow(image, norm=LogNorm(vmin=5))
```
* Inspect the spectrum, and try to locate the regions of the power spectrum that contain the signal and those which contain the periodic noise. Use array slicing to set the noise regions to zero.
* Apply the inverse Fourier transform to plot the resulting image.

````python
from matplotlib.colors import LogNorm
from matplotlib import cm

# noised image.
fig =  plt.figure(figsize = (15, 7))
image = plt.imread('moonlanding.png')
plt.imshow(image, plt.cm.gray)
plt.title('Original Image');
````

````python

image_fft = fftpack.fft2(image)
def spectrum(image_fft):
    power = np.abs(image_fft)
    plt.imshow(power, plt.cm.gray, norm=LogNorm(vmin=5))
    plt.colorbar()

plt.figure(figsize = (10, 5))
spectrum(image_fft)
plt.title('2D Frequency of Fourier transform');
````

````python
# Filter Spectrum in FFT
k= 0.08
image_fft2 = image_fft.copy()
# Set r and c to be the number of rows and columns of the array.
r, c = image_fft2.shape
image_fft2[int(r*k):int(r*(1-k))] = 0
image_fft2[:, int(c*k):int(c*(1-k))] = 0
plt.figure(figsize = (10, 5))
spectrum(image_fft2)
plt.title('Filtered Spectrum');
````

````python
# Reconstruct the denoised image from the filtered spectrum.
image_new = fftpack.ifft2(image_fft2).real

plt.figure(figsize = (15, 7))
plt.imshow(image_new, plt.cm.gray)
plt.title('2D Filtered Image in FFT');
````
